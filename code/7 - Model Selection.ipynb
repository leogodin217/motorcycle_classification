```python
# Python setup
%reload_ext autoreload
%autoreload 2
%matplotlib inline

import torch.nn as nn
#import data functions and constants (no caps for easier typing)
import modelingfunctions.dataprocessing as d
import modelingfunctions.modeling as m
import modelingfunctions.utilities as u
```

## Resnet models

While Resnet18 learns more quickly, we see that after fifty epochs, both Resnet34 and Resnet50 outperform the simpler model. As an added bonus, all three models take about the same time to train. It appears that copying data to cuda takes most of the training time. 

```python
num_targets = 50
batch_size = 64
epochs=50
```

## Resnet-18

```python
dataloaders = d.get_target_dataloader(d.basic_data, num_targets=num_targets,
                                      batch_size=batch_size)
model = m.get_model(num_targets, generalization='batchnorm', layers='18')
optimizer = m.get_adam_optimizer(model)
criterion = nn.CrossEntropyLoss()

results = m.train_model(dataloaders=dataloaders, model=model, criterion=criterion,  topk=3,
                        name="Resnet-18", optimizer=optimizer, epochs=epochs, 
                        clear_cuda_cache=True, print_epoch=True)
print(results['model_run_time'])
u.plot_results(results)
u.plot_accuracy(results)
```
