```python
from google_images_download import google_images_download
from pathlib import Path
import re
```

```python
# Download some data from 2017
response = google_images_download.googleimagesdownload()
#Usig specific names in multiple requests so we likely get multiple images of each motorcycle
paths = response.download({'keywords': '2017 kawasaki', 'limit': 100, 'output_directory': '../data/motorcycle_data', 'specific_site': 'totalmotorcycle.com'})
paths = response.download({'keywords': '2017 suzuki', 'limit': 100, 'output_directory': '../data/motorcycle_data', 'specific_site': 'totalmotorcycle.com'})
paths = response.download({'keywords': '2017 harley', 'limit': 100, 'output_directory': '../data/motorcycle_data', 'specific_site': 'totalmotorcycle.com'})
paths = response.download({'keywords': '2017 honda', 'limit': 100, 'output_directory': '../data/motorcycle_data', 'specific_site': 'totalmotorcycle.com'})
paths = response.download({'keywords': '2017 indian', 'limit': 100, 'output_directory': '../data/motorcycle_data', 'specific_site': 'totalmotorcycle.com'})
paths = response.download({'keywords': '2017 triumph', 'limit': 100, 'output_directory': '../data/motorcycle_data', 'specific_site': 'totalmotorcycle.com'})
paths = response.download({'keywords': '2017 ducati', 'limit': 100, 'output_directory': '../data/motorcycle_data', 'specific_site': 'totalmotorcycle.com'})
paths = response.download({'keywords': '2017 yamaha', 'limit': 100, 'output_directory': '../data/motorcycle_data', 'specific_site': 'totalmotorcycle.com'})
```

```python
# Get a list of file names for processing
# Download could be easier by specifying a specific directory instead of using the search
# terms. Since data is already downloaded, We will work with it.
base_path = Path('./motorcycle_data')
paths = []
folders = [x for x in base_path.iterdir() if x.is_dir()]
for folder in folders:
    paths = paths + [x.name for x in folder.iterdir()]
```

```python
def get_motorcycle_class(name):
    # Everything between the image number and .jpg
    base_name = re.match(r'\d+\. ([\w-]+)\.jpg', name).group(1)
    # Remove -small
    no_small = re.sub('-small', '', base_name)
    # Remove image size
    no_size = re.sub(r'-\d{3, 4}x\d{3, 4}', '', no_small)
    # Remove the index identifier (ducati-monster3 or ducati-monsterb)
    final_name = re.sub(r'\w$', '', no_size)
    return final_name
    
```

```python
classes = []

for path in paths:
    classes.append(get_motorcycle_class(path))
   
```
